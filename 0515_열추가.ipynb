```python
data1 = pd.read_csv('성매매광고트윗.csv', encoding = 'euc_kr')
data1
```

```python
data1['tweet']
```

### 트윗 길이 열 추가

```python
tweet_length = []
for i in range(len(data1)):
    tweet_length.append(len(data1['tweet'][i]))
```

```python
data1['tweet_len'] = tweet_length
```

### 닉네임 길이 열 추가

```python
nick_length = []
for i in range(len(data1)):
    nick_length.append(len(data1['name'][i]))
```

```python
data1['nick_len'] = nick_length
```

### 닉네임 내 물음표 개수

```python
nick_countQ = []
for i in range(len(data1)):
    nick_countQ.append(data1['name'][i].count('?'))
```

```python
data1['nick_count_Q'] = nick_countQ
```

### 트윗 내 물음표 개수

```python
tweet_countQ = []
for i in range(len(data1)):
    tweet_countQ.append(data1['tweet'][i].count('?'))
```

```python
data1['tweet_count_Q'] = tweet_countQ
```

### 트윗 내 공백 개수

```python
tweet_countSpace = []
for i in range(len(data1)):
    tweet_countSpace.append(data1['tweet'][i].count(' '))
```

```python
data1['tweet_count_space'] = tweet_countSpace
```

### 트윗 내 공백 개수/트윗 길이

```python
arr1 = pd.array(tweet_length)
arr2 = pd.array(tweet_countSpace)
```

```python
data1['space_div_len'] = (arr2 /arr1).round(3)
data1[:5]
```

```python
data1.to_csv('성매매광고트윗_트렌드a.csv', index=False, encoding="utf-8-sig")
```
